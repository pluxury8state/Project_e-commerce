```python
# 5. Используя pandas, проведи когортный анализ пользователей.
# В период с января по декабрь выяви когорту с самым высоким retention на 3й месяц.
```

```python
# чтобы посчитать retention я посчитал нужным написать функцию, которая будет принимать 
# нужный датафрейм, а также диапазон временного ряда для анализа
```

```python
def get_retention(df: pd.DataFrame, start_cagort_date: str, end_cagort_date: str):
    """
    Функция get_retention предназначена для подсчета retention пользавателей
    
    :param df: датафрейм, по которому будет происходить подсчет retention (тип данных -> pd.DataDrame)
    :param start_cagort_date:  дата начала когортного анализа в виде Год-Месяц-День  (тип данных -> str)
    :param end_cagort_date:  дата окончания когортного анализа в виде Год-Месяц-День (тип данных -> str)
    """
    
    # переводим даты начала и конца когортного анализа в DateTime
    start_date = datetime.datetime.strptime(start_cagort_date, "%Y-%m-%d")
    end_date = datetime.datetime.strptime(end_cagort_date, "%Y-%m-%d")
    
    # Урезаем данные от начала до конца анализа 
    stripped_df = df.query("order_purchase_timestamp >= @start_date and order_purchase_timestamp <= @end_date")
    
    # Урезаем дату покупки заказа до формата: Год-Месяц
    stripped_df['order_purchase_timestamp'] = pd.to_datetime(stripped_df.order_purchase_timestamp.dt.strftime('%Y-%m'))
    
    # Формируем датафрэйм с когортами
    cohorts = (stripped_df.groupby(['order_purchase_timestamp', 'customer_unique_id']).
           agg({'order_id': 'count'})
           .drop(columns='order_id')
           .reset_index()
           .drop_duplicates(subset='customer_unique_id'))
    
    # Формируем вспомоательный датафрейм, в котором есть вся информация о покупателях и их покупках
    all_perchaces_and_customers = (stripped_df.groupby(['order_purchase_timestamp', 'customer_unique_id'])
                     .agg({'order_id': 'count'})
                     .drop(columns='order_id')
                     .reset_index())

    # Сформируем словарь когорт с уникальными id покупателей
    cagorts_dict = {}
    for date in list(cohorts.order_purchase_timestamp.unique()):
        cagorts_dict[date] = list(cohorts.query("order_purchase_timestamp == @date").customer_unique_id)

    # Сформируем словарь ключей когорт, чтобы было легче итерироваться
    cagorts_dict_keys_list = list(cagorts_dict.keys())
    
    retention_df = pd.DataFrame() # создаем пустой датафрейм

    for i in range(len(cagorts_dict_keys_list)):  # получаем индексы массива
        current_cagort_time = cagorts_dict_keys_list[i] # текущее время когорты
        current_cagort_users = cagorts_dict[current_cagort_time] # все пользователи текущей кагорты

        retention_list = [] # список retantion нужен для создания DataFrame

        for j in range(i, len(cagorts_dict_keys_list)): # получаем индексы для каждой итерации
            next_cagort_time = cagorts_dict_keys_list[j] # текущая или следующая дата когорты

            # пользователи купившие в следующую дату
            all_users_in_next_date = list(all_perchaces_and_customers.query("order_purchase_timestamp == @next_cagort_time")["customer_unique_id"])

            # подсчитываем retention через sets
            retention = len(set(current_cagort_users).intersection(set(all_users_in_next_date))) * 100 / len(current_cagort_users)
            # print(cagorts_dict_keys_list[i], next_cagort_time, "-", retention)

            #добавляем полученный retention в словарь
            retention_list.append(retention)

        # записываем данные в датафрейм
        row = pd.DataFrame(data=retention_list, columns=[pd.to_datetime(current_cagort_time)]) # делаем датафрейм-строку

        retention_df = pd.concat([retention_df, row], ignore_index=False, axis=1) # соединяем подсчеты

    retention_df = retention_df.transpose() # переворачиваем табличку
    
    return retention_df
```

```python
df = orders_data.merge(users_data, on="customer_id") # делаем объединение,
                                                                    # чтобы установить связь заказах и клиентах
start_cagort_date = "2017-01-01" #дата начала анализа
end_cagort_date = "2017-12-31" # дата конца анализа

# передаем арнументы в функцию
retantion_data = get_retention(df, start_cagort_date, end_cagort_date)

# подкрасим значения
styled_df = (retantion_data
            .style
            .set_caption('User retention by cohort')
            .background_gradient(cmap='viridis')
            .highlight_null('white')
            .format("{:.2f}%",na_rep=""))
styled_df
```

```python
# # Задание 6. 
# Часто для качественного анализа аудитории использую подходы, основанные на сегментации.
# Используя python, построй RFM-сегментацию пользователей, чтобы качественно оценить свою аудиторию.
# В кластеризации можешь выбрать следующие метрики: R - время от последней покупки пользователя до текущей даты,
# F - суммарное количество покупок у пользователя за всё время, M - сумма покупок за всё время.
# Подробно опиши, как ты создавал кластеры. Для каждого RFM-сегмента построй границы метрик recency,
# frequency и monetary для интерпретации этих кластеров.
# Пример такого описания: RFM-сегмент 132 (recency=1, frequency=3, monetary=2)
# имеет границы метрик recency от 130 до 500 дней, frequency от 2 до 5 заказов в неделю,
# monetary от 1780 до 3560 рублей в неделю.
```

```python
# Для начала неплохо было бы составить более удобную таблицу, чтобы мы по ней смогли построить RFM-сегментацию пользователей
```

```python
# создадим датафрэйм с данными о количестве товаров в заказе и их стоимости
order_items_price = products_data.drop(columns=['product_id', 'seller_id', 'shipping_limit_date', 'freight_value'])
order_items_price.head()
```

```python
# создадим датафрэйм с данными о номере заказа, id покупателя, датой создания заказа и датой оплаты заказа
orders = orders_data.drop(columns=['order_status',
                                 'order_delivered_carrier_date',
                                 'order_delivered_customer_date',
                                 'order_estimated_delivery_date'])

orders.head()
```

```python
# создадим датафрэйм с данными об уникальных id клиентов
customers = users_data.drop(columns=['customer_zip_code_prefix',
                                       'customer_city',
                                       'customer_state'])
customers.head()
```

```python
# Объединим все датафреймы в один
rfm_analysis = pd.merge(order_items_price, orders, on="order_id").merge(customers, on="customer_id")

# переименуем колонку order_item_id и удалим ненужную колонки с псевдо id клиента
rfm_analysis = rfm_analysis.rename(columns={"order_item_id": "item_count"}).drop(columns=["customer_id"])
rfm_analysis.head()
```

```python
# проверка на пропуски 
rfm_analysis.isna().sum()
```

```python
# для начала избавимся от неоплаченных заказов
rfm_analysis = rfm_analysis.loc[~rfm_analysis.order_approved_at.isna()]
# проверка на пропуски 
rfm_analysis.isna().sum()
```

```python
# посчитаем доход от оплаченых заказов
rfm_analysis["total_price"] = rfm_analysis["item_count"] * rfm_analysis["price"]
rfm_analysis.head()
```

```python
# посчитаем итоговую цену каждого заказа
total_sum_of_orders = rfm_analysis.groupby(["customer_unique_id", "order_id"], as_index=False).agg({"total_price": "sum"})
total_sum_of_orders.head()
```

```python
# считаем средний чек пользователя
mean_revenue_from_users = total_sum_of_orders.groupby("customer_unique_id", as_index=False) \
                                        .agg({"total_price": "mean"}) \
                                        .rename(columns={"total_price": "total_mean_price"})
mean_revenue_from_users.head()
```

```python
# Посчитаем частоту заказов каждого пользователя за весь период
frequency_df = rfm_analysis.groupby("customer_unique_id", as_index=False) \
            .agg({"order_id": "count"}) \
            .rename(columns={"order_id": "frequency"})
frequency_df.head()
```

```python
# Посчитаем время от последней покупки пользователя до текущей даты для каждого пользователя.
# Будем считать разницу в днях между датой последнего заказа в наших данных и датой последнего заказа конкретного пользователя 

# найдем дату последнего заказа в данных
last_date = rfm_analysis.order_purchase_timestamp.max()
```

```python
# сформируем датафрейм с последними датами заказов пользователей
duration_df = rfm_analysis.groupby("customer_unique_id", as_index=False).agg({"order_purchase_timestamp": "max"})
duration_df.head()
```

```python
# Зададим колонку с последней датой заказов в данных
duration_df["last_date"] = last_date
duration_df.head()
```

```python
# посчитаем разницу между последними покупками пользователей и последней датой покупок в данных
duration_df["recency"] = duration_df["last_date"] - duration_df["order_purchase_timestamp"]
duration_df.head()
```

```python
# Создадим функции для разделения данных на 3 класса
# Для recency: 1 - давние, 2 - средней давности, 3 - недавние
# Для frequency: 1 - разовые, 2 - редкие, 3 - частые
# Для total_mean_price: 1 - маленький средний чек, 2 - средняя величина чека, 3 - большой средний чек
# Для деления на классы будем использовать следующую логику: разобьем данные о показателях на 3 квантиля,
# чтобы равномерно распределить оценку
```

```python
# получим кванитили распределения для total_mean_price
money_33_quntile = mean_revenue_from_users.total_mean_price.quantile(1 / 3.0)
money_66_quntile = mean_revenue_from_users.total_mean_price.quantile(2 / 3.0)
```

```python
# получим квантили распределения для frequency
```

```python
frequency_33_quntile = frequency_df.frequency.quantile(1 / 3.0)
frequency_66_quntile = frequency_df.frequency.quantile(2 / 3.0)
```

```python
print(frequency_33_quntile, frequency_66_quntile)
```

```python
# как видим с квантилями частоты заказов такой трюк не провернуть,
# тогда берем максимальное и минимальное значиние частоты, вычитаем из
# и делим на три части
```

```python
one_part_of_frequency = (frequency_df.frequency.max() - frequency_df.frequency.min()) / 3.0
one_part_of_frequency
```

```python
# получим квантили распределения для recency

recency_33_quntile =  duration_df["recency"].quantile(1/3.0)
recency_66_quntile =  duration_df["recency"].quantile(2/3.0)
```

```python
# функция для нахождения класса получаемых денег
def get_class_by_total_mean_price(value, first_quantile, second_quntile):
    if value <= first_quantile:
        return 1
    elif value > first_quantile and value <= second_quntile:
        return 2
    else:
        return 3
```

```python
# функция для нахождения класса частоты
def get_class_by_frequency(value, first_part, second_part):
    if value <= first_part:
        return 1
    elif value > first_part and value <= second_part:
        return 2
    else:
        return 3
```

```python
# функция для нахождения класса давности покупки
def get_class_by_recency(value, first_part, second_part):
    if value <= first_part:
        return 1
    elif value > first_part and value <= second_part:
        return 2
    else:
        return 3
```

```python
# перед примением функций объединим наши датафреймы
r_f_m_df = duration_df.merge(frequency_df, on="customer_unique_id") \
            .merge(mean_revenue_from_users, on="customer_unique_id")
# удалим ненужные колонки 
r_f_m_df = r_f_m_df.drop(columns=["last_date", "order_purchase_timestamp"])

r_f_m_df.tail()
```

```python
# считаем R
r_f_m_df["R"] = r_f_m_df \
            .apply(lambda row: get_class_by_recency(row[1], recency_33_quntile, recency_66_quntile), axis=1)
```

```python
# считаем F
r_f_m_df["F"] = r_f_m_df \
            .apply(lambda row: get_class_by_total_mean_price(row[2], one_part_of_frequency, one_part_of_frequency * 2), axis=1)
```

```python
# считаем M
r_f_m_df["M"] = r_f_m_df \
            .apply(lambda row: get_class_by_total_mean_price(row[3], money_33_quntile, money_66_quntile), axis=1)
```

```python
r_f_m_df["segment"] = r_f_m_df.apply(lambda x: str(x[4]) + str(x[5]) + str(x[6]), axis=1)
```
